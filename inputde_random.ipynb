```python
# By using commas
age = int(input())
name = input()
print('My name is',name,'and my Age is',age,'years old')
```

```python
# Using modulo operator
age = int(input())
name = input()
print('My name is %s and age is %d years old'%(name,age))
```

```python
# By using dot format method
age = int(input())
name = input()
print('My name is {} and my Age is {} years old'.format(age,name))
```

```python
# By using f-format method
age = int(input())
name = input()
print(f'My name is {name} and my Age is {age} years old')
```

```python
# By using f-format method
age = 20.436637
name = 'pavan'
print(f'My name is {name} and my Age is {age:.2f} years old')
```
