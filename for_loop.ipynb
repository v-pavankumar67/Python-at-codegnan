```python
# for loop
n = int(input())
for num in range(1,n+1):
    print(f"{num} - {num**2}")
print("program done")
```

```python
# Print 1 to n numbers
n = int(input())
for num in range(1,n+1):
    print(num)
```

```python
# Print 1 to n numbers to print in horizantally using end=' '
n = int(input())
for num in range(1,n+1):
    print(num,end=' ')
```

```python
# To print list of even and odd numbers between n1&n2
n1=int(input())
n2=int(input())
even=[]
odd=[]
for num in range(n1,n2+1):
    if num%2:
        num.append(odd)
    else:
        num.append(even)
print(even)
print(odd)
```
