````python
#set allows only unique elements
#set is unordered collection
s = {10,2,5,'hi',"hello",2,5,'a'}
print(s)
````

````python
# set does not support indexing
s = {10,2,5,'hello',"hi",2}
print(s[0])
````

````python
#### SET METHODS
- add(val): It adds the value to set of the value is already not exists in set
- update(new_set): It adds all new set elements
- remove(val): It removes the specified value from set if it is present in set,
               otherwise ```KeyError```
- Discard(val): It removes the value if it is present in set,otherwise nothing.
- union(new_set): It returns the all unique elements from sets.
- intersection(new_set): It returns teh common elements from all sets.
- Difference(new_set): It returns the  set elements those are not present in new_set.
- Symmetric_difference(new_set): It returns the common elements from the sets.
````

````python
s = {2,6,'hi',4,9,"hello"}
print(s)
````

````python
#Add value
s = {2,6,'hi',4,9,"hello"}
s.add(10)
print(s)
````

````python
# Update the new set
s = {2,6,'hi',4,9,"hello"}
print(s)
s.update({11,24,"pavan"})
print(s)
````

````python
# Remove(val),discard(val)
s = {2,6,'hi',4,9,"hello"}
s.remove(6)
print(s)
s.remove(6) # It raises key error due to dictionary key

````

````python
s = {2,6,'hi',4,9,"hello"}
s.discard(2)
print(s)
s.discard(2) # It wont raise any error.
print(s)
````

````python
# To find union of sets
a = {1,2,3}
b = {4,3,2,6}
union = a.union(b)
print(union)
print(a | b) #union symbol
#intersection 
print(a.intersection(b))
print(a & b)
# Difference
print(a.difference(b))
print(a - b)
# Symmetric difference
print(a.symmetric_difference(b))
print(a ^ b)
#TO print subset
print(a.issubset({1,2,3,4}))
# superset
print({1,2,3,4}.issuperset(a))
# Disjoint sets
print({1}.isdisjoint(b))
print(a,b)






````
